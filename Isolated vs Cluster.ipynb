```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
h = fits.open('RedMapperCluster_1_Cluster')
i = fits.open('RedMapperCluster_1_Iso')
j = fits.open('RedMapperCluster_1_Group')
```

```python
h.info()
i.info()
j.info()
```

```python
H = h[1].data
I = i[1].data
J = j[1].data

Radius_Clu = H.field(4)
Prob_Clu = H.field(5)
Weight_Clu = H.field(8)*H.field(9)
Radius_Iso = I.field(4)
Prob_Iso = I.field(5)
Weight_Iso = I.field(8)*I.field(9)
Radius_Gro = J.field(28)
Prob_Gro = J.field(29)
Weight_Gro = J.field(32)*J.field(33)

print(Radius_Iso)
```

```python
plt.figure(1).add_axes((0,0,1.2,1.2))

plt.scatter(Prob_Iso, Radius_Iso, marker ='s', s=10, color = 'blue', label = "Isolated")
plt.scatter(Prob_Clu, Radius_Clu, marker ='s', s=10, color = 'red', label = "Cluster")
plt.scatter(Prob_Gro, Radius_Gro, marker ='s', s=10, color = 'Orange', label = "Group")

plt.xlabel("Membership Probability")
plt.ylabel("Radius From Cluster Centre ($Mpc$)")
plt.legend(loc = 'lower left')

print(len(Radius_Iso)-len(Radius_Gro))
print(len(Radius_Gro))
print(len(Radius_Clu))
```

```python
plt.figure(1).add_axes((0,0,1.2,1.2))

plt.scatter(Prob_Iso, Weight_Iso, marker ='s', s=10, color = 'blue', label = "Isolated")
plt.scatter(Prob_Clu, Weight_Clu, marker ='s', s=10, color = 'red', label = "Cluster")
plt.scatter(Prob_Gro, Weight_Gro, marker ='s', s=10, color = 'Orange', label = "Group")

plt.xlabel("Membership Probability")
plt.ylabel("Luminosity and Radial Weight")
plt.legend(loc = 'lower right')
```
