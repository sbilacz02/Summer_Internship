```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Spirals = np.array([7,7,7,7,9,9,12,28,34,24,36,37,27])
Ellipticals = np.array([4,4,5,1,4,7,4,39,8,9,16,5,9])
Lenticulars = np.array([4,2,8,8,4,8,7,7,26,19,19,14,37])

S_Frac = np.array([7/15,7/16,9/17,9/24,12/23,30/74,34/68,24/52,36/71,37/56])
E_Frac = np.array([4/15,4/16,4/17,7/24,4/23,26/74,8/68,9/52,16/71,5/56])
L_Frac = np.array([4/15,5/16,4/17,8/24,7/23,18/74,26/68,19/52,19/71,14/56])

Richness = np.array([5.1451, 24.3969, 46.0132, 68.5601, 90.2221, 111.320, 134.919, 152.325, 168.160, 185.901])
```

```python
from scipy.optimize import curve_fit

def func(x, a, b, c):
    return a*x + b*x**2 + c

popt, pcov = curve_fit(func, Richness, S_Frac)
print(popt)

popt2, pcov2 = curve_fit(func, Richness, E_Frac)
print(popt2)

popt3, pcov3 = curve_fit(func, Richness, L_Frac)
print(popt3)

plt.figure(1).add_axes((0,0,1.2,1.2))

plt.plot(Richness, func(Richness, *popt), color = 'blue', linewidth = '1')
plt.plot(Richness, func(Richness, *popt2), color = 'red', linewidth = '1')
plt.plot(Richness, func(Richness, *popt3), color = 'orange', linewidth = '1')

plt.scatter(Richness, S_Frac, color = 'blue', label = 'Spirals', marker = 'x')
plt.scatter(Richness, E_Frac, color = 'red', label = 'Ellipticals', marker = 'o')
plt.scatter(Richness, L_Frac, color = 'orange', label = 'Lenticulars/Unknowns', marker = 's')

plt.xlabel("Richness")
plt.ylabel("Fraction of Population")
plt.legend(loc = "upper left")
plt.title("Morphology vs Richness")
```
